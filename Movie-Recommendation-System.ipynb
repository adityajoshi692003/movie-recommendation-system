```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
m['genres'] = m['genres'].apply(convert)
```

```python
def convert1(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
m['cast'].apply(convert1)
```

```python
m['cast'] = m['cast'].apply(convert1)
```

```python
def fetch_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
m['crew'].apply(fetch_crew)
```

```python
m['crew'] = m['crew'].apply(fetch_crew)
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y = ps.stem(i)
    return " ".join(y)
```

```python
def recommend(m):
    movie_index = new_df[new_df['title'] == m].index[0]
    dist = similarity[movie_index]
    movies_list = sorted(list(enumerate(dist)), reverse = True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```
